```python
from sitkva.models import *
from sitkva import db
```

```python
streets = {}

for fl in Street.query.all():
    streets[fl.name] = {"latitude" : fl.latitude, "longitude" : fl.longitude}
```

```python
for fl in FlatRemodel.query.filter(FlatRemodel.latitude != None).all():
    fl.latitude = None
    fl.longitude = None
    db.session.add(fl)
    db.session.commit()
    print(fl, " added")
```
